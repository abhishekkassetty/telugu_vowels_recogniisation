```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import pathlib
import sys
import cv2
from keras.models import Sequential,Model
from keras_preprocessing.image import ImageDataGenerator
from keras.layers import Dense,Conv2D,Flatten,Dropout,Activation,BatchNormalization,MaxPooling2D,GlobalAveragePooling2D
from keras import regularizers,optimizers
from keras.losses import CategoricalCrossentropy
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from keras.callbacks import ModelCheckpoint
```

```
train_path = r"/content/Final Dataset of Telugu Handwritten Chararcters/Test1/achulu"
```

```
categories = ["a","aa","ah","ai","am","ao","e","ee","i","ii","o","oo","ru","ruu","u","uu"]
labels = [i for i in range(len(categories))]
label_dict = {categories[i]:labels[i] for i in range(len(categories))}
```

```
img = cv2.imread("/content/Final Dataset of Telugu Handwritten Chararcters/Test1/achulu/a/1.jpg")
```

```
img.shape
```

```
plt.imshow(img)
```

```
data = []
target = []
```

```
for category in categories:
    filepath = os.path.join(train_path,category)
    img_names=os.listdir(filepath)
    for img_name in img_names:
        img_path=os.path.join(filepath,img_name)
        img=cv2.imread(img_path)
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        resized=cv2.resize(gray,(50,50))
        data.append(resized)
        target.append(label_dict[category])
        
```

```
data = np.array(data)
data = data/255
data.shape
```

```
data = np.reshape(data,(data.shape[0],50,50,1))
```

```
target = np.array(target)
target.shape
```

```
target
```

```
new_target = to_categorical(target)
```

```
new_target.shape
```

```
new_target
```

```
new_target
```

```
model = Sequential()
model.add(Dense(200,input_shape = data.shape[1:],activation = "relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Conv2D(150,(3,3),activation = "relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Flatten())
model.add(Dense(150,activation = "relu"))
model.add(Dense(100,activation = "relu"))
model.add(Dense(16,activation = "softmax"))
```

```
model.compile(loss = "categorical_crossentropy",optimizer = "adam",metrics = ["accuracy"])
```

```

checkpoint = ModelCheckpoint("model-{epoch:03d}.model",save_best_only=True,mode="auto")
```

```
fitting = model.fit(data,new_target,validation_split=0.2,epochs = 15,callbacks=[checkpoint])
```
